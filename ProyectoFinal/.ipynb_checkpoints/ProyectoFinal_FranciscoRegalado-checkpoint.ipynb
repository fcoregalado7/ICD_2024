## 1. Introduccin


### 1.1 Sobre la base de datos

La base de datos consiste en la recopilacin de imagenes de granos de cafe con distintos tostdos. Estos tostdos estan clasificados en *Tostado Bajo*, *Tostado Medio*, *Tostado Oscuro* y sin tostr, es decir, *Granos en verde*. De esta forma, la base de datos se divide en 4 clases: `Dark`, `Green`, `Light` y `Medium`.Estas cuatro clases indican el nivel de tostado de los granos de caf.

* Los granos de cafe de las clases `Green` y `Light` pertenecen a la varietal _Laos Typica Bolaven_.
* Los granos de cafe de la clase `Medium` pertenecen a la varietal _Doi Chaang_.
* Mientras que los granos de cafe de la clase `Dark` pertenecen a la varietal _Brazil Cerrado_.

Existen 400 imagenes por clase, dando un total de 1600 imagenes con un tamao de 224x224.

La base de datos fue extraida de [Kaggle](https://www.kaggle.com/datasets/gpiosenka/coffee-bean-dataset-resized-224-x-224/data) de los autores Gerry, Sroison, Sakdipat y Thitaree.

### 1.2 Objetivo

El objetivo es mostrar como se comparan los mtodos de clasificacin estadisticos contra los mtodos de redes neuronales para clasificar los distintos tostdos en los granos de caf.

### 1.3 Motivacin

El consumo del cafe ha formado parte de la historia moderna de la humanidad. El caf se ha adaptado a los consumidores por su alta variabilidad en los espectros de sabores que se pueden desarrollar en el cultivo de plantas de cafe y en los procesos de extraccin de los granos de cafe de sus respetivos frutos, creandooslos una gran diversida de variedades de plantas con su propio perfil de identidad (intesidad, notas, aromas, etc.). Los tostdores de caf utilizan tecnicas para estandarizar sus tostdos y garantizar un perfil de sabor. Una variable fundamental para garantizar estos estandares es la llamada curva de tostado que se refiere a la variabilidad de la temperatura utilizada en el tostdo durante el transcurso de tiempo. Ajustar esta curva permite obtener los distintos perfiles de tostado. 

Tener diferentes tipos de tostdo permite acceder a la variabilidad de perfiles de sabor que se pueden extraer del grano de cafe. Mientras un tostdo bajo resalta ms los sabores herbales del grano, un tostdo medio rescata los sabores frutales y dulces del caf. En contraste, un tostado oscuro da sabores mas intensos y amargos producto de la alta combustion que se le dio al grano durante el tostdo.

No hay un perfil que sea mejor que el otro, si no que cada perfil se adapta a las necesidades del consumidor. Por esta razn, tener un clasificador eficiente de los tipos de tostado en los granos de caf resulta atractivo e interesante con el fin de garantizar una distribucin correcta de los granos de caf tostdo.

```python
data_dir = '/Users/paco/Documents/GitHub/ICD_2024/ProyectoFinal/Coffee'
categories = ['Dark', 'Green', 'Light', 'Medium']

# Inicilizamos las listas para imgenes y etiquetas

images = []
labels = []
```

#### 3.3 Procesamiento de datos

```python
# Carga y preprocesamiento de imgenes
for category in categories:
    folder_path = os.path.join(data_dir, category)
    for img_name in os.listdir(folder_path):
        try:
            img_path = os.path.join(folder_path, img_name)
            img = imread(img_path)
            img_resized = resize(img, (128, 128))  # Redimensionar las imgenes a 128x128
            images.append(img_resized)
            labels.append(category)
        except Exception as e:
            print(f'Error al cargar la imagen {img_name}: {e}')

# Convertir listas a arrays de NumPy
X = np.array(images)
y = np.array(labels)
```

#### 4.1 VGGNet

Red neuronal que procesa las imagenes paso a paso de manera secuencial. Cada capa toma la salida de la anterior y pasa el resultado a la siguiente. Piensa en una serie de filtros que analizan diferentes detalles de la imagen (bordes, texturas, formas) capa por capa. Las capas finales se conectan de manera ms "clsica" para combinar todos estos detalles y dar una clasificacin (ChatGPT).

Hiperparametros

```python
num_conv_layers = 5 
num_filters = [64, 128, 256, 512, 512]  
dropout_rate = 0.5  
learning_rate = 0.001 
weight_decay = 1e-4  
optimizer_choice = 'adam'
epochs = 50
batch_size = 64
num_classes = 4
train_split = 0.7
validation_split = 0.2
```

Arquitectura VGGNet

```python
class CustomVGG(nn.Module):
    def __init__(self, num_conv_layers, num_filters, dropout_rate):
        super(CustomVGG, self).__init__()
        
        layers = []
        in_channels = 3  # Cambiar a 1 para imgenes en escala de grises
        
        # Aadir las capas convolucionales
        for i in range(num_conv_layers):
            layers.append(nn.Conv2d(in_channels, num_filters[i], kernel_size=3, padding=1))
            layers.append(nn.ReLU(inplace=True))
            layers.append(nn.MaxPool2d(kernel_size=2, stride=2))
            in_channels = num_filters[i]
            
        self.conv_layers = nn.Sequential(*layers)
        
        # Aadir capa completamente conectada (temporalmente) hasta conocer el tamao de la salida de las capas convolucionales
        self.fc_layers = None
        
        self.dropout_rate = dropout_rate

    def forward(self, x):
        x = self.conv_layers(x)
        x = x.view(x.size(0), -1)  # Aplanar para las capas completamente conectadas
        
        if self.fc_layers is None:  # Si las capas FC no estn definidas, definirlas en base a las dimensiones de x
            self.fc_layers = nn.Sequential(
                nn.Linear(x.size(1), 4096),  # Calcula dinmicamente el tamao de entrada
                nn.ReLU(inplace=True),
                nn.Dropout(self.dropout_rate),
                nn.Linear(4096, 4096),
                nn.ReLU(inplace=True),
                nn.Dropout(self.dropout_rate),
                nn.Linear(4096, 10)  # Para una clasificacin en 10 clases (puedes cambiarlo)
            ).to(x.device)
        
        x = self.fc_layers(x)
        return x
```

Creamos el modelo

```python
model = CustomVGG(num_conv_layers, num_filters, dropout_rate)
```

Optimazador y perdidas

```python
if optimizer_choice == 'adam':
    optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=weight_decay)
elif optimizer_choice == 'sgd':
    optimizer = optim.SGD(model.parameters(), lr=learning_rate, weight_decay=weight_decay, momentum=0.9)

# Def loss
criterion = nn.CrossEntropyLoss()
```

Transformacin de los datos

```python
transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,)) ])

# Leemos los datos previamente capturados en data_dir
dataset = datasets.ImageFolder(root=data_dir, transform=transform)
```

Split Base de datos

```python
train_size = int(train_split * len(dataset))
val_size = int(validation_split * len(dataset))
test_size = len(dataset) - train_size - val_size
train_dataset, val_dataset, test_dataset = random_split(dataset, [train_size, val_size, test_size])
```

Se leen los datos

```python
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
validate_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
class_names = ['Dark', 'Green', 'Light', 'Medium']
```

Definimos las mtricas

```python
def calculate_metrics(output, target):
    pred = torch.argmax(output, dim=1)
    f1 = f1_score(target.cpu().numpy(), pred.cpu().numpy(), average='macro')
    recall = recall_score(target.cpu().numpy(), pred.cpu().numpy(), average='macro')
    accuracy = accuracy_score(target.cpu().numpy(), pred.cpu().numpy())
    return f1, recall, accuracy
```

```python
train_loss_history = []
val_loss_history = []
train_f1_history = []
val_f1_history = []
train_recall_history = []
val_recall_history = []
train_accuracy_history = []
val_accuracy_history = []
start_time = time.time()
```

Entrenamiento y Validacin

```python
def train_and_validate(model, train_loader, validate_loader, criterion, optimizer, epochs):

    
    for epoch in range(epochs):
        # Entrenamiento
        model.train()
        train_loss = 0.0
        train_f1 = 0.0
        train_recall = 0.0
        train_accuracy = 0.0
        num_batches = 0

        for data, target in train_loader:
            data, target = data.to(device), target.to(device)
            optimizer.zero_grad()
            output = model(data)
            loss = criterion(output, target) / np.log(num_classes)
            loss.backward()
            optimizer.step()

            train_loss += loss.item()

            # Calcular mtricas
            f1, recall, accuracy = calculate_metrics(output, target)
            train_f1 += f1
            train_recall += recall
            train_accuracy += accuracy
            num_batches += 1

        train_loss /= num_batches
        train_f1 /= num_batches
        train_recall /= num_batches
        train_accuracy /= num_batches
        
        train_loss_history.append(train_loss)
        train_f1_history.append(train_f1)
        train_recall_history.append(train_recall)
        train_accuracy_history.append(train_accuracy)

        # Validacin
        model.eval()
        val_loss = 0.0
        val_f1 = 0.0
        val_recall = 0.0
        val_accuracy = 0.0
        num_batches = 0

        with torch.no_grad():
            for data, target in validate_loader:
                data, target = data.to(device), target.to(device)
                output = model(data)
                loss = criterion(output, target)

                val_loss += loss.item()

                # Calcular mtricas
                f1, recall, accuracy = calculate_metrics(output, target)
                val_f1 += f1
                val_recall += recall
                val_accuracy += accuracy
                num_batches += 1

        val_loss /= num_batches
        val_f1 /= num_batches
        val_recall /= num_batches
        val_accuracy /= num_batches
        
        val_loss_history.append(val_loss)
        val_f1_history.append(val_f1)
        val_recall_history.append(val_recall)
        val_accuracy_history.append(val_accuracy)

        print(f'Epoch {epoch+1}/{epochs} - '
              f'Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, '
              f'Train F1: {train_f1:.4f}, Val F1: {val_f1:.4f}, '
              f'Train Recall: {train_recall:.4f}, Val Recall: {val_recall:.4f}, '
              f'Train Accuracy: {train_accuracy:.4f}, Val Accuracy: {val_accuracy:.4f}')

    
    return train_loss_history, val_loss_history, train_f1_history, val_f1_history, train_recall_history, val_recall_history, train_accuracy_history, val_accuracy_history
```

Device

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

Llamamos a funcin de entrenamiento y validacin

```python
train_loss_history, val_loss_history, train_f1_history, val_f1_history, train_recall_history, val_recall_history, train_accuracy_history, val_accuracy_history = train_and_validate(model, train_loader, validate_loader, criterion, optimizer, epochs)
```

Definimos la Matriz de Confusin

```python
def plot_confusion_matrix(model, loader):
    all_preds = []
    all_labels = []
    model.eval()
    with torch.no_grad():
        for data, target in loader:
            data, target = data.to(device), target.to(device)
            output = model(data)
            preds = torch.argmax(output, dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(target.cpu().numpy())
    
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Etiqueta verdadera')
    plt.ylabel('Etiqueta predicha')
    plt.title('Matriz de confusin Red Neuronal VGGNet')
    plt.show()
```

#### 4.2 DenseNet
 Se enfoca en reutilizar toda la informacin que ha ido "aprendiendo" a lo largo del camino. Aqu, cada capa no solo toma lo que le da la anterior, sino que tambin tiene acceso directo a todas las capas previas. Es como si cada nueva capa tuviera acceso a toda la "sabidura" acumulada por las capas anteriores, permitiendo un flujo de informacin mucho ms rico (ChatGPT).

Hiperparametros

```python
learning_rate = 0.001
growth_rate = 12
block_config=(6, 12, 24, 16)
bn_size=4
batch_size = 10
dropout_rate=0.2
num_epochs = 50
num_classes = 4 
train_split = 0.7
validation_split = 0.2
```

Arquitectura DenseNet

```python
class DenseLayer(nn.Module):
    def __init__(self, in_channels, growth_rate, bn_size, dropout_rate):
        super(DenseLayer, self).__init__()
        inter_channels = bn_size * growth_rate
        self.bn1 = nn.BatchNorm2d(in_channels)
        self.conv1 = nn.Conv2d(in_channels, inter_channels, kernel_size=1)
        self.bn2 = nn.BatchNorm2d(inter_channels)
        self.conv2 = nn.Conv2d(inter_channels, growth_rate, kernel_size=3, padding=1)
        self.relu = nn.ReLU(inplace=True)
        self.dropout = nn.Dropout(p=dropout_rate)

    def forward(self, x):
        out = self.conv1(self.relu(self.bn1(x)))
        out = self.dropout(out)
        out = self.conv2(self.relu(self.bn2(out)))
        out = torch.cat([x, out], 1)
        return out

class Transition(nn.Module):
    def __init__(self, in_channels, out_channels, dropout_rate):
        super(Transition, self).__init__()
        self.bn = nn.BatchNorm2d(in_channels)
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size=1)
        self.relu = nn.ReLU(inplace=True)
        self.pool = nn.AvgPool2d(2)
        self.dropout = nn.Dropout(p=dropout_rate)

    def forward(self, x):
        out = self.pool(self.relu(self.conv(self.bn(x))))
        return out

class DenseNet(nn.Module):
    def __init__(self, num_init_features, growth_rate, block_config, bn_size, num_classes, dropout_rate):
        super(DenseNet, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, num_init_features, kernel_size=7, stride=2, padding=3, bias=False),
            nn.BatchNorm2d(num_init_features),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=3, stride=2, padding=1),
        )

        num_features = num_init_features
        for i, num_layers in enumerate(block_config):
            block = nn.Sequential(*(DenseLayer(num_features + j * growth_rate, growth_rate, bn_size, dropout_rate) for j in range(num_layers)))
            self.features.add_module(f'denseblock{i+1}', block)
            num_features += num_layers * growth_rate
            if i != len(block_config) - 1:
                trans = Transition(num_features, num_features // 2, dropout_rate)
                self.features.add_module(f'transition{i+1}', trans)
                num_features //= 2

        self.features.add_module('bn_norm', nn.BatchNorm2d(num_features))
        self.classifier = nn.Sequential(
            nn.Linear(num_features, num_classes),
            nn.Dropout(p=dropout_rate)
        )

    def forward(self, x):
        features = self.features(x)
        out = F.adaptive_avg_pool2d(features, (1, 1))
        out = out.view(out.size(0), -1)
        out = self.classifier(out)
        return out
```

Transformar los datos

```python
# Transformaciones
transformations = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor(),
    transforms.Normalize([0.485], [0.229])
])
```

Leer los datos

```python
dataset = datasets.ImageFolder(root=data_dir, transform=transformations)
train_size = int(train_split * len(dataset))
val_size = int(validation_split * len(dataset))
test_size = len(dataset) - train_size - val_size
train_dataset, val_dataset, test_dataset = random_split(dataset, [train_size, val_size, test_size])
```

```python
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

Modelo de Instancias

```python
model = DenseNet(num_init_features=16, growth_rate=growth_rate, block_config=block_config, 
                 bn_size=bn_size, num_classes=num_classes, dropout_rate=dropout_rate).to(device)
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
criterion = nn.CrossEntropyLoss()
```

Metricas

```python
train_losses = []
val_losses = []
train_accuracies = []
val_accuracies = []
train_recalls = []
val_recalls = []
train_f1_scores = []
val_f1_scores = []
start_time = time.time()
```

Entrenamiento y Validacin

```python
for epoch in range(num_epochs):
    # Entrenamiento
    model.train()
    running_loss = 0.0
    all_labels_train = []
    all_preds_train = []
    
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        _, predicted = torch.max(outputs.data, 1)
        all_labels_train.extend(labels.cpu().numpy())
        all_preds_train.extend(predicted.cpu().numpy())
        
        running_loss += loss.item()
    
    train_losses.append(running_loss / len(train_loader))
    train_accuracies.append(accuracy_score(all_labels_train, all_preds_train))
    train_recalls.append(recall_score(all_labels_train, all_preds_train, average='macro'))
    train_f1_scores.append(f1_score(all_labels_train, all_preds_train, average='macro'))
    
    # Validacin
    model.eval()
    running_val_loss = 0.0
    all_labels_val = []
    all_preds_val = []
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels) / np.log(num_classes)
            _, predicted = torch.max(outputs.data, 1)
            all_labels_val.extend(labels.cpu().numpy())
            all_preds_val.extend(predicted.cpu().numpy())
            running_val_loss += loss.item()
    
    val_losses.append(running_val_loss / len(val_loader))
    val_accuracies.append(accuracy_score(all_labels_val, all_preds_val))
    val_recalls.append(recall_score(all_labels_val, all_preds_val, average='macro'))
    val_f1_scores.append(f1_score(all_labels_val, all_preds_val, average='macro'))

    print(f'Epoch {epoch + 1}/{num_epochs}, Train Loss: {train_losses[-1]:.4f}, Val Loss: {val_losses[-1]:.4f}, '
          f'Train Acc: {train_accuracies[-1]:.4f}, Val Acc: {val_accuracies[-1]:.4f}')
```

#### 6 Discusin

En los mtodos estadisticos, utilizar PCA muestra mejores resultados que trabajar con HOG.
PCA (Anlisis de Componentes Principales) es una tcnica que reduce la dimensinalidad de los datos al identificar las direcciones (componentes principales) que capturan la mayor varianza en el conjunto de datos. En este contexto de imagenes de caf tostdo, las diferencias entre las clases Oscuro, Verde, Claro y Medio, probablemente en variaciones globales de color y textura. PCA puede ser efectivo para capturar estas variaciones globales, ya que considera la informacin de todos los pixeles y puede rescatar diferencias sutiles en tonalidad y luminosidad. 

HOG (Histogramas de Gradientes Orientados) se enfoca en capturar caractersticas locales relacionadas con la forma y los bordes en una imagen, analizandooslo las orientaciones de los gradientes en pequeas regiones. Sin embargo, en las imgenes de granos de caf tostdo las diferencias entre las clases no restan realmente representadas por bordes o formas distintivas, si no ms bien por patrones de color y texturas. Adems, HOG convierte las imgenes en escala de grises perdiendo informacin importante para la clasificacin.

Para estos mtodos, SVC obtuvo el mayor rendimiento. Con un F1-Score de 0.96 y Precisin de 0.96 con una validacin cruzada de 0.96 utlizando PCA. Mientras que, utilizandoselo HOG, SVC obtuvo un F1-Score de 0.82 y Precisin de 0.76 con una validacin cruzada de 0.72. Por lo que SVC utilizandoselo PCA es el mtodo con mejores resultados de los clasificadores estadsticos.

En comparacin, las Redes Neuronales demostraron un rendimiento superior en tanto F1-Score y Precisin respecto a los mtodos estadsticos. VGGNet, en su ultima epoca alcanzo un 0.99 de precisin y 0.99 de F1-Score. DenseNet obtiene nmeros cercanos en su ejecucin. VGGNet resulta ligeramente mejor que DenseNet en general. Ambas demuestran un rendimiento muy alto clasificando las imagenes. 

Evaluando los tiempos de ejecucin en los mtodos estadsticos con PCA, estos no pasarn de 2 minutos en la ejecucin de los 4 modelos conjuntos. Utilizando HOG, la ejecucin de los 4 modelos conjuntos alcanzo los 10 minutos, sin embargo no estuvo ni cercano a los resultados obtenidos mediante PCA. Pero estos quedan muy por atras en cuando costo computacional utilizado por las redes neuronales. Para ambas los tiempos de ejecucin fueron mayores de 20 minutos para este conjunto de entrenamiento a 50 epocas cada una. En promedio, el tiempo de entrenamiento fue de 25 minutos para DenseNet y 36 minutos para VGGNet. Se alcanzo un muy alto rendimiento pero a un costo computacional mucho mayor en contraste con los mtodos estadsticos. 

El tiempo computacional se reduciria dismunyendo la cantidad de epocas. Segun las grficas parece ser suficiente con 20 epocas, lo que disminuira el tiempo de entrenamiento para las redes neuronales.
