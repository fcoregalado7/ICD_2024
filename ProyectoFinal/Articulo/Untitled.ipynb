```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Datos de ejemplo
data = {
    'Clase': ['SVC', 'SVC', 'DT', 'DT', 'KNN', 'KNN', 'RF', 'RF'],
    'Conjunto': ['PCA', 'HOG', 'PCA', 'HOG', 'PCA', 'HOG','PCA', 'HOG'],
    'Valores': [0.96, 0.82, 0.81, 0.66, 0.81, 0.63, 0.93, 0.76]
}

# Convertir los datos a un DataFrame de pandas
df = pd.DataFrame(data)

# Crear un gráfico de barras con seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x='Clase', y='Valores', hue='Conjunto', data=df)

# Configurar etiquetas y título
plt.title('Precisión de Métodos Estadisticos')
plt.xlabel('Métodos Estadísticos')
plt.ylabel('Precisión')

# Guardar la gráfica en formato PNG
plt.savefig('PrecisionColectiva.png', format='png')

# Mostrar la gráfica
plt.show()
```
